```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston


boston = load_boston()
print(boston.data.shape)
```

```python
boston.DESCR
```

```python
y = boston.target
x_raw =boston.data
x=pd.DataFrame(x_raw, columns = ['CRIM', 'ZN', 'TNDUS', 'CHAS', 'NOX' , 'RM' , 'AGE','DIS','RAD','TAX','a','b','c'])
x=x.loc[:,'CRIM':'NOX']
    
```

```python
x
```

```python
plt.hist(x['CRIM'], bins='auto')
plt.xlabel("CRIM")
plt.ylabel("Frequency")
plt.title('Histogram of CRIM (per capita crime rate by town)')
plt.show()
```

```python
for i in x.columns:
    plt.boxplot(x[i])
    plt.xlabel(i)
    plt.ylabel("Frequency")
    plt.title('Box plot of {}'.format(i))
    plt.show()
```

```python
def correlation_matrix(df):
    from matplotlib import pyplot as plt
    from matplotlib import cm as cm
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    cmap = cm.get_cmap('jet', 30)
    cax = ax1.imshow(df.corr(), interpolation="nearest", cmap=cmap)
    plt.title('Abalone Feature Correlation')
    labels=["aa"]
    for x in df.columns.tolist():
        labels.append(x)
    print(labels)
    ax1.set_xticklabels(labels,fontsize=6)
    ax1.set_yticklabels(labels,fontsize=6)
    # Add colorbar, make sure to specify tick locations to match desired ticklabels
    fig.colorbar(cax, ticks=[.75,.8,.85,.90,.95,1])
    plt.show()

correlation_matrix(x)
```

```python
plt.scatter(x['TNDUS'],x['NOX'])
plt.xlabel("TNDUS")
plt.ylabel('NOX')
plt.title('Scatter plot of TNDUS and NOX')
plt.show()
```

```python
from sklearn.decomposition import PCA
pca=PCA()
x_transformed = pca.fit_transform(x)
x_transformed=pd.DataFrame(x_transformed,columns=['First principal Component','Second principal Component','Third principal Component','Fourth principal Component','Fifth principal Component'])
x_transformed
```

```python
plt.scatter(x_transformed.loc[:,'First principal Component'],x_transformed.loc[:,'Second principal Component'])
plt.xlabel('First principal Component')
plt.ylabel('Second principal Component')
plt.title('Scatter plot of First principal Component and Second principal Component')
plt.show()
```

```python
plt.scatter(x_transformed.loc[:,'Fourth principal Component'],x_transformed.loc[:,'Fifth principal Component'])
plt.xlabel('Fourth principal Component')
plt.ylabel('Fifth principal Component')
plt.title('Scatter plot of Fourth principal Component and Fifth principal Component')
plt.show()
```
